```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# 1. Label Encoding
label_encoder = LabelEncoder()
df['koi_disposition'] = label_encoder.fit_transform(df['koi_disposition'])
df['koi_pdisposition'] = label_encoder.fit_transform(df['koi_pdisposition'])
```

```python
# 2. Feature Selection
# Select features likely to be relevant for classification
features = [
    'koi_score', 'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec', 
    'koi_period', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 
    'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad'
]
X = df[features]
y = df['koi_disposition']  # Target variable: 'koi_disposition' (0 = FP, 1 = Candidate, 2 = Confirmed)
```

```python
# 3. Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 4. Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Dictionary to store model performances
model_performance = {}
```

```python
# Function to train, predict, and evaluate a model
def train_evaluate_model(model, model_name):
    # Train the model
    model.fit(X_train, y_train)
    # Make predictions
    y_pred = model.predict(X_test)
    # Evaluate model
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    print(f"\nModel: {model_name}\nAccuracy: {accuracy:.4f}")
    print("Classification Report:\n", report)
    model_performance[model_name] = accuracy
```

```python
# 1. Logistic Regression
logreg = LogisticRegression(max_iter=1000)
train_evaluate_model(logreg, "Logistic Regression")
```

```python
# 2. Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
train_evaluate_model(rf, "Random Forest")
```

```python
# 3. Support Vector Machine
svm = SVC(kernel='rbf', random_state=42)
train_evaluate_model(svm, "SVM")
```

```python
# 4. Gradient Boosting Classifier
gbc = GradientBoostingClassifier(random_state=42)
train_evaluate_model(gbc, "Gradient Boosting")
```

```python
# Train a Random Forest model on the entire training data to extract feature importances
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Extract feature importances
feature_importances = rf_model.feature_importances_
features = X.columns  # Get feature names
```

```python
# Create a DataFrame for visualization
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False).head(10)  # Top 10 features
```

```python
# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df, x='Importance', y='Feature', palette="viridis")
plt.title('Top 10 Feature Importances in Random Forest Model')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Convert X_train and X_test to DataFrames for easier column selection by name
X_train_df = pd.DataFrame(X_train, columns=features)  # Ensure feature_columns is the list of original column names
X_test_df = pd.DataFrame(X_test, columns=features)
```

```python
# Select the top 10 features based on importance
selected_features = importance_df['Feature'].values  # Get the feature names as an array
X_train_selected = X_train_df[selected_features]  # Use column names to select features
X_test_selected = X_test_df[selected_features]    # Same for the test set
```

```python
# Define parameter grid for tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Implement grid search for hyperparameter tuning
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_selected, y_train)
```

```python
# Best parameters and accuracy
print("Best Parameters:", grid_search.best_params_)
print("Best Training Accuracy:", grid_search.best_score_)
```

```python
# Evaluate on test set
best_rf_model = grid_search.best_estimator_
test_accuracy = best_rf_model.score(X_test_selected, y_test)
print("Test Accuracy with Best Parameters:", test_accuracy)
```

```python
def recommend_habitable_exoplanets(model, data, top_n=10):
    """
    Recommends the top N habitable exoplanets based on the trained model, koi_score, and other criteria.
    
    Parameters:
    - model: Trained machine learning model (e.g., Random Forest)
    - data: DataFrame with features used for prediction
    - top_n: Number of exoplanets to recommend (default is 10)
    
    Returns:
    - DataFrame containing the top N habitable exoplanets with filtered results.
    """
    # Ensure only the top 10 features are used
    top_features = importance_df['Feature'].values
    data_selected = data[top_features]

    # Ensure features are scaled using the same scaler
    data_scaled = scaler.transform(data_selected)

    # Predict classes using the model
    predictions = model.predict(data_scaled)

    # Add predictions as a new column to the original DataFrame
    data['Predicted_Disposition'] = predictions

    # Filter for habitable exoplanets (Class 2: Confirmed)
    habitable_exoplanets = data[data['Predicted_Disposition'].isin([2, 1])]

    # Apply additional filtering for habitability
    habitable_exoplanets = habitable_exoplanets[
        (habitable_exoplanets['koi_score'] >= 0.9) &
        (habitable_exoplanets['koi_prad'] >= 0.5) & (habitable_exoplanets['koi_prad'] <= 2.0) &  # Earth-like radius
        (habitable_exoplanets['koi_teq'] >= 200) & (habitable_exoplanets['koi_teq'] <= 400) &  # Habitable temperature
        (habitable_exoplanets['koi_insol'] >= 0.5) & (habitable_exoplanets['koi_insol'] <= 2.0)  # Insolation flux near Earth's value
    ]

    # Sort by koi_score (higher is better)
    habitable_exoplanets = habitable_exoplanets.sort_values(by='koi_score', ascending=False)

    # Select the top N exoplanets
    top_habitable = habitable_exoplanets.head(top_n)
    top_habitable= top_habitable[['kepid','kepoi_name','koi_score', 'koi_prad', 'koi_teq', 'koi_period', 'koi_insol','Predicted_Disposition']]
    top_habitable_renamed = top_habitable.rename(columns={
        'kepid':'Kepler ID',
        'kepoi_name':'Name',
        'koi_score': 'Habitability Score',
        'koi_prad': 'Planet Radius',
        'koi_teq': 'Equi.Temperature (K)',
        'koi_period': 'Orbital Period',
        'koi_insol': 'Insolation Flux',
        'Predicted_Disposition': 'Disposition'
    })

    # Return key details of the top exoplanets
    return top_habitable_renamed
```

```python
# Apply the function to recommend the top 10 habitable exoplanets
top_10_exoplanets = recommend_habitable_exoplanets(best_rf_model, df, top_n=10)
print("Top 10 Habitable Exoplanets:")
top_10_exoplanets
```
